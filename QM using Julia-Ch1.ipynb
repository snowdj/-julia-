## 1.1 概要

为什么经济学家要选择Julia编程？

1、诺贝尔奖得主Sargent在推荐（https://julia.quantecon.org/https://julia.quantecon.org/）；

2、美联储在使用(https://libertystreeteconomics.newyorkfed.org/2015/12/the-frbny-dsge-model-meets-julia.htmlhttps://libertystreeteconomics.newyorkfed.org/2015/12/the-frbny-dsge-model-meets-julia.html)

3、经济学家们都说它好，例如，JFV et al.（2018）、Coleman et al.(2017)

4、Julia的主要优势在于结合了Python、Matlab的高级语言的自然句法和Fortran、C/C++等底层语言的速度

5、从旧编程语言转换到Julia的成本较小，因为Julia语法命令与Matlab基本一致，且可以很容易地嵌入Fortran/C/C++代码
